```python
import pandas as pd
import scipy.stats as stats
def chi_square_test(data, col1, col2):
    contingency_table = pd.crosstab(data[col1], data[col2])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    return chi2, p, dof, expected
```

```python
# Perform chi-square test between 'Price_Range' and 'Location'
# 'df' is the DataFrame that holds the loaded CSV data
chi2, p, dof, expected = chi_square_test(df, 'Price_Range', 'Location')

print("Chi-Square Value:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequency Table:\n", expected)
```

```python
import numpy as np
import scipy.stats as stats
```

```python
# Update observed values based on function-based contingency table
observed = np.array([[0, 0, 7],  # High
                     [2, 3, 0],  # Low
                     [1, 3, 4]]) # Medium
```

```python
# Compute the row and column sums
row_sums = observed.sum(axis=1, keepdims=True)
col_sums = observed.sum(axis=0, keepdims=True)
total = observed.sum()
```

```python
# Compute expected frequencies
expected = (row_sums @ col_sums) / total
```

```python
# Compute Chi-Square statistic
chi2_stat = ((observed - expected) ** 2 / expected).sum()
```

```python
# Degrees of Freedom = (rows - 1) * (columns - 1)
dof = (observed.shape[0] - 1) * (observed.shape[1] - 1)
```

```python
# Compute p-value
p_value = 1 - stats.chi2.cdf(chi2_stat, dof)

print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)
print("Expected Frequency Table:\n", expected)
```
